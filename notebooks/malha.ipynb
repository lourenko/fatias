### Traçar um gráfico simples

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi

# Dados
Nx = 16
x = np.linspace(-pi, pi, Nx) # -pi até pi
y = np.sin(x)

# Canvas
fig, ax = plt.subplots()

# Plots
ax.plot(x, y, '0.35')
```

```python
class Malha:
    "Representa uma malha 2D"

    def __init__(self, nx, ny, l, h, ax,\
                 c='0.35'):
        "Construtor"
        self.Nx = nx # Número de nós em x
        self.Ny = ny # Número de nós em y
        self.L = l # Comprimento da malha
        self.H = h # Altura da malha
        self.cor = c # Cinza padrão
        self.ax = ax
        
    def plota(self):
        "Apresenta a malha"
        lx = [0.0, self.L]
        ly = [0.0, self.H]
        dy = self.H/(self.Ny-1)
        dx = self.L/(self.Nx-1)
        # Primeiro as linhas horizontais
        for j in range(self.Ny):
            Dy =[j*dy, j*dy]
            self.ax.plot(lx, Dy, self.cor)
        # Então as linhas vertical
        for i in range(self.Nx):
            Dx =[i*dx, i*dx]
            self.ax.plot(Dx, ly, self.cor)
```

```python
fig, ax = plt.subplots()
m = Malha(32, 24, 1.8, 1.2, ax, c='C1')
m.plota()
```

```python
m2 = Malha(16, 24, 0.8, 2.2, ax, c='C4')
m2.plota()
plt.show()
```
